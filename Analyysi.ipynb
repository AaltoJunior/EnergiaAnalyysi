```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("estat_nrg_cb_pem_en.csv")
```

```python
print(data["siec"].unique())
```

```python
data.loc[data["siec"] == "RA300"]["geo"].unique()
```

```python
fi = data.loc[data["geo"] == "FI"]
fi.loc[:,'TIME_PERIOD'] = pandas.to_datetime(fi['TIME_PERIOD']) 
```

```python
plt.figure(figsize=(16,10))
suomi = plt.plot(fi.loc[fi["siec"] == "C0000"].TIME_PERIOD, fi.loc[fi["siec"] == "C0000"].OBS_VALUE, color="r")

plt.gca().set(title="RA500_5160 Suomessa", xlabel="Pivmr",ylabel="GWH")
plt.legend((suomi),("GWH"))
```
